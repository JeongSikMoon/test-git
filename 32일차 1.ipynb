```python
NUM_WORDS = 10000
```

```python
(X_train,y_train),(X_test,y_test) = tf.keras.datasets.imdb.load_data(num_words=NUM_WORDS)
```

```python
def multi_hot_sequences(sequences, dimension):
    # 0으로 채워진 (len(sequences), dimension) 크기의 행렬을 만듭니다
    results = np.zeros((len(sequences), dimension))
    for i, word_indices in enumerate(sequences):
        results[i, word_indices] = 1.0  # results[i]의 특정 인덱스만 1로 설정합니다
    return results
```

```python
X_train = multi_hot_sequences(X_train, dimension=NUM_WORDS)
y_test = multi_hot_sequences(y_test, dimension=NUM_WORDS)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(16, activation='relu',input_shape=(NUM_WORDS,)),  #2가지 방식 tf.nn.relu=상수(객체),'relu'=문자
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid') 
])
```
