```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
msc = MinMaxScaler()
```

```python
X = msc.fit_transform(iris.iloc[:,:-1])
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
y = le.fit_transform(iris.iloc[:,-1])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
y = ohe.fit_transform(iris[['species']]).toarray()
```

```python
y
```

```python
def create_model():
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Dense(16,activation=tf.nn.relu,input_shape=(4,)))
    model.add(tf.keras.layers.Dense(3,activation=tf.nn.softmax))
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['acc'])
    return model
```

```python
kc = KerasClassifier(create_model) #요거가 핵심 Scikit-Learn <-> keras 함수집어넣는다. 함수뒤에 ()안붙힌다. 
```

```python
kc.fit(iris.iloc[:,:-1],iris.iloc[:,-1])
```

```python
kc.fit(X,y, epochs=10) #fit도 같다.
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(kc,X,y,cv=10) #y를 one-hot encoding
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(kc,{'optimizer':['rmsprop','adam']})
```
