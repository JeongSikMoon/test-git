```python
from sklearn.datasets import make_classification
```

```python
data = make_classification(1000,5) #랜덤하게 연습용 데이터 만들어준다.
```

```python
data
```

```python
d = pd.DataFrame(data[0])
dt = pd.DataFrame(data[1],columns=['target'])
```

```python
d
```

```python
data = pd.concat([d,dt],axis=1)
```

```python
l = learning_curve(LogisticRegression(),data.iloc[:,:-1],data.iloc[:,-1],cv=10)
```

```python
plot.learning_curve(l[1],l[2],l[0]) #LogisticRegression
```

```python
k = learning_curve(KNeighborsClassifier(),data.iloc[:,:-1],data.iloc[:,-1],cv=10)
```

```python
plot.learning_curve(k[1],k[2],k[0]) #KNeighborsClassifier
```
