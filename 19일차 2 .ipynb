```python
import matplotlib.pyplot as plt
```

```python
plt.figure() #가상의 영역 여러 옵션을 넣고 싶을때 사용
```

```python
plt.style.available  #matplotlib에서 사용할 수 있는 기본스타일 종류
```

```python
plt.style.use('grayscale')
```

```python
#plt.axes(ylim=(0,50)) #기본 그래프를 그려준다. axes안써도 그래프를 만들어 준다.
# plt.figure(figsize=(20,5),facecolor=('red')) #figsize:전체크기를 바꾼다.
                                             #facecolor:순서가 필요하다.
plt.title('Title') #제목
plt.xlabel('Xlabel') #X축 라벨
plt.ylabel('Ylabel') #Y축 라벨
plt.grid(True) #그리드를 그려준다.
#plt.axis('off')
# plt.grid(axis='x') #그리드 x만 나오게한다.
# plt.grid(axis='y') #그리드 y만 나오게한다.
plt.plot([1,2,3],[4,5,6],color = 'blue',marker='+',label='xlabel') #('bo')blue point
plt.plot([4,5,6],[1,2,3],color = 'red',marker='o',label='ylabel') #('bo')blue point
#plt.figure(figsize=(20,5),facecolor=('red')) #figsize:전체크기를 바꾼다.
                                             #facecolor:순서가 필요하다.
plt.legend()    
```
