```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(8) #PCA를 여러 값으로 바꿈
```

```python
 t = pca.fit_transform(wine.iloc[:,:-1])
```

```python
x = pd.DataFrame(t)
```

```python
x
```

```python
wine_pca = pd.concat([x,wine.target],axis=1)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
cross_val_score(KNeighborsClassifier(),wine.iloc[:,:-1],wine.iloc[:,-1],cv=10).mean()
```

```python
cross_val_score(KNeighborsClassifier(),wine_pca.iloc[:,:-1],wine_pca.iloc[:,-1],cv=10).mean()
```
