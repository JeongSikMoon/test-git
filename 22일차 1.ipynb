```python
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset('iris')
```

```python
pd.get_dummies(data.species)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit_transform(data.species)
```

```python
x = pd.DataFrame(le.fit_transform(data.species), columns=['target'])
```

```python
x
```

```python
data_pd = pd.concat([data,x],axis = 1) #iris 데이터 뒤에 합친다.
```

```python
data_pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(data_pd[data_pd.columns[:-2]], data_pd.target)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test) #정확도를 나타낸다.
```

```python
from sklearn.model_selection import cross_val_score
```

```python
r = KNeighborsClassifier()
```

```python
c = cross_val_score(r,data_pd[data_pd.columns[:-2]],data_pd.target,cv=10)
```

```python
temp=[]
for i,j in enumerate(range(20)):
    r = KNeighborsClassifier(i+3)
    c = cross_val_score(r,data_pd[data_pd.columns[:-2]],data_pd.target,cv=10)
    temp.append(np.mean(c))
```
